```python
import sqlite3
import pandas as pd
from pandas import DataFrame
```

```python
conn = sqlite3.connect('UFC_database')  
c = conn.cursor()
```

```python
fighter_physical_stats_df = pd.read_csv('fighter_physical_stats_csv')
fighter_career_stats_df = pd.read_csv('fighter_career_stats_csv')
events_df = pd.read_csv('event_df_csv')
event_chart_stats = pd.read_csv('chart_stats_csv')
```

```python
fighter_physical_stats_df.to_sql('Physical_Stats', conn, if_exists='append', index='False')
fighter_career_stats_df.to_sql('Career_Stats', conn, if_exists='append', index='False')
events_df.to_sql('Events', conn, if_exists='append', index='False')
event_chart_stats.to_sql('Fight_Chart_Stats',conn,  if_exists='append', index="False")
```

```python
Phys_Stats = pd.DataFrame(c.execute("SELECT * FROM Physical_Stats").fetchall())
Phys_Stats.drop([0], axis=1, inplace=True)
Phys_Stats.columns = ['Name', 'Wins', 'Loses', 'Draw', 'No_Contests', 'Height', 'Weight', 'Reach', 'Stance', 'DOB']
Phys_Stats.head()
```

```python
Career_Stats = pd.DataFrame(c.execute("""SELECT * FROM Career_Stats""").fetchall())
Career_Stats.drop([0], axis=1, inplace=True)
Career_Stats.columns = ["Name", "SLpMs", "StrAccs", "SApMs", "StrDefs", "TDAvgs", "TDAccs", "TDDefs", "SubAvgs"]
Career_Stats.head()
```

```python
Fighter_Profiles = pd.concat([Phys_Stats, Career_Stats], keys=['Name'])
```
