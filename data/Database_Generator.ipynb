```python
import sqlite3
import pandas as pd
from pandas import DataFrame
```

```python
conn = sqlite3.connect('UFC_Fighter_Encyclopedia.db')  
c = conn.cursor()
```

```python
fighter_physical_stats_df = pd.read_csv('fighter_physical_stats_csv')

fighter_physical_stats_df.to_sql('Physical_Stats', conn, if_exists='append', index='False')
```

```python
df = pd.DataFrame(c.execute('SELECT * from Physical_Stats').fetchall())
df
```

```python
fighter_basic_stats_df = pd.read_csv('fighter_basic_stats_csv')

fighter_basic_stats_df.to_sql('Basic_Stats', conn, if_exists='append', index='False')
```

```python
df = pd.DataFrame(c.execute('SELECT * from Basic_Stats').fetchall())
df
```

```python
fighter_career_stats_df = pd.read_csv('fighter_career_stats_csv')

fighter_career_stats_df.to_sql('Career_Stats', conn, if_exists='append', index='False')
```

```python
df = pd.DataFrame(c.execute('SELECT * from Career_Stats').fetchall())
df
```
