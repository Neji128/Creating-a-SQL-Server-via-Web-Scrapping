```python
import sqlite3
import pandas as pd
from pandas import DataFrame
```

```python
conn = sqlite3.connect('UFC_database')  
c = conn.cursor()
```

```python
fighter_physical_stats_df = pd.read_csv('fighter_physical_stats_csv')
fighter_career_stats_df = pd.read_csv('fighter_career_stats_csv')
events_df = pd.read_csv('event_df_csv')
```

```python
fighter_physical_stats_df.to_sql('Physical_Stats', conn, if_exists='append', index='False')
fighter_career_stats_df.to_sql('Career_Stats', conn, if_exists='append', index='False')
events_df.to_sql('Events', conn, if_exists='append', index='False')
```

```python
df = pd.DataFrame(c.execute('SELECT * from Physical_Stats').fetchall())
df.head()
```
